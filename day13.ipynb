```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('robot 1.jpg')
img2 = cv2.imread('robot 2.jpg')
img = img[0:360,0:540,:]
```

```python
img.shape , img2.shape
```

```python
def showing(im):
    plt.imshow('Preview',im)
    plt.show()
def imshow(im):
    cv2.imshow('Preview',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def showing(im):
    plt.imshow(im[:,:,::-1])
    plt.show()
```

```python
imshow(img)
imshow(img2)
```

```python
showing(img)
```

```python
showing(img2)
```

```python
added = cv2.add(img,img2)
showing(added)
```

```python
import numpy as np
```

```python
img2_1 = cv2.multiply(img2,2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
## Adusting Contrast (alpha) and Brightness (beta)
img2_2 = cv2.convertScaleAbs(img2,alpha = 0.5,beta = 10)
showing(img2_2)
```

```python
img2_3 = cv2.resize(img2,(400,400), interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
showing(img2_3)
```

```python
img2_3 = cv2.resize(img2,None, fx=2,fy=0.5,interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
showing(img2_3)
```
